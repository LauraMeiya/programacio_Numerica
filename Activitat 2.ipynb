```python
column_rellevants = delayed_flights[['Year','Month','DepTime','CRSDepTime','ArrTime','CRSArrTime','ArrDelay','Distance','Origin','AirTime']]
```

```python
# Hi ha 6019 dades faltants en la columna ArrTime

# Afegir una columna

column_rellevants["Temps vol"] = column_rellevants["ArrTime"] - column_rellevants["DepTime"]
column_rellevants.head(10)
```

```python
# Em pensava que restant el temps d'arrivada i de sortida sortiria el temps del vol, perÃ² hi ha valors negatius.

def retard(x):
    if x>=0:
        return "Perdoneu pel retard"
    else:
        return "Bon viatge"

column_rellevants["Retard"] = column_rellevants["ArrDelay"].apply(retard)
column_rellevants.head(10)
```

```python
# Velocitat mitjana del vol

column_rellevants["Velocitat"] = column_rellevants["Distance"] / column_rellevants["AirTime"]
column_rellevants.head(10)
```

```python
# No he trobat columna d'aerolÃ­nies, perÃ² el aeroport que mÃ©s ha tingut retards ha estat ATL.
# Vols mÃ©s llargs

vols_llargs = column_rellevants[['AirTime']]
vols_llargs_final = vols_llargs.sort_values(by="AirTime", ascending=False)
vols_llargs_final.head()
```
